# 필요 라이브러리 임포트

```python
import cv2
import matplotlib.pyplot as plt
```

# 원본 이미지 보기

```python
src = cv2.imread('data/cat.jpg')
src_cvt = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
plt.imshow(src_cvt)
```

# 마스킹 이미지 만들기
0이 아닌 픽셀값 : 내가 필요한 부분  
0인 픽셀값 : 없애겠다

## 원본 이미지에서 고양이 부분만 마스킹할 수 있도록 특정 영역만 자른다
cv2.inRange(): 이미지의 최소 rgb, 최대 rgb 색상인 것만 가져오겠다

```python
# src의 최소 범위, 최대 범위 이 범위 안만 자르겠다(배경색인 흰색 빼고 다 가져오겠다.)
mask = cv2.inRange(src, (0,0,0), (253,253,253)) # b,g,r 순

# 이미지 저장
# cv2.imwrite('data/cat_mask.jpg', mask) # 파일명

plt.imshow(mask, cmap='gray')
plt.show()
```

# 배경 이미지

```python
dst = cv2.imread('data/redcarpet.jpg')
dst_cvt = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)

plt.imshow(dst_cvt)
```

## 이미지 사이즈 맞추기

```python
# height, width, channel
src.shape
```

```python
dst_resize = cv2.resize(dst, (src.shape[1], src.shape[0]))
print(dst.shape, dst_resize.shape)
dst_resize_cvt = cv2.cvtColor(dst_resize, cv2.COLOR_BGR2RGB)
plt.imshow(dst_resize_cvt)
```

# 마스킹

```python
cv2.copyTo(src_cvt, mask, dst_resize_cvt)

plt.imshow(dst_resize_cvt)
```
